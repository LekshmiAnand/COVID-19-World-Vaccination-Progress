```python
df_case = df_case[['Country/Region','Germany']].copy()
df_case.head()          #rename columns
```

```python
df_case.columns=['date','cases']           #rename columns
```

```python
#convert object  to date
df_case['date'] =  pd.to_datetime(df_case['date'], format='%m/%d/%y')
df_case
```

```python
#copy date and data from selected country
df_vacc = df_vacc.loc[df_vacc['country']=='Germany']
df_vacc
```

```python
#convert object  to date
df_vacc['date'] =  pd.to_datetime(df_vacc['date'], format='%Y-%m-%d')
```

```python
#merge dataset vaccination progress and dataset from cases in Germany
df = pd.merge(df_case,df_vacc, how='left' , left_on='date', right_on='date')
df
```

```python
df.drop(columns=['Unnamed: 0'],inplace = True)
```

```python
df.columns
```

```python
df = df[['date','cases','people_vaccinated','people_fully_vaccinated','daily_vaccinations']].copy()
```

```python
#Replace NaN Values with Zeros
df['people_vaccinated'] = df['people_vaccinated'].fillna(0)
df['people_fully_vaccinated'] = df['people_fully_vaccinated'].fillna(0)
df.tail(10)
```

```python
df.to_csv(os.path.join(path,'PreparedData', 'case_vaccination_combined.csv'))
```

## Create a correlation heatmap using seaborn: 

```python
# Create a subset excluding the "Date" columns

sub = df[[ 'cases','people_vaccinated','people_fully_vaccinated','daily_vaccinations']]
```

```python
# Create a subplot with matplotlib
f,ax = plt.subplots(figsize=(10,10))

# Create the correlation heatmap in seaborn by applying a heatmap onto the correlation matrix and the subplots defined above.
corr = sns.heatmap(sub.corr(), annot = True, ax = ax)
```

The higher the daily vaccination the lower the cases. They are negatively correlated

```python
fig = corr.get_figure()
fig.savefig(os.path.join(path, 'Visualization','heatmap.png'))
```

```python
# Create a scatterplot for the "daily_vaccinations" and "total_vaccinations" columns in seaborn

plt= sns.lmplot( x = 'daily_vaccinations',y = 'cases', data = df)
```

```python
#Exporting map as png file
plt.savefig(os.path.join(path, 'Visualization','scatter.png'))
```

As the vaccination increases ther cases decreases tremendiusly.
There is a very negative relationship, the trend line is very steep.
